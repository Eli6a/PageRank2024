```python
import pandas as pd
import plotly.express as px
import re
from ipywidgets import interact, FloatRangeSlider
```

```python

# Fonction pour lire le fichier et extraire les données
def load_data(filepath):
    data = []
    with open(filepath, 'r') as file:
        for line in file:
            match = re.match(r"\('<http://dbpedia.org/resource/(.*?)>', (.*?)\)", line.strip())
            if match:
                link, score = match.groups()
                data.append((link, float(score)))
    return pd.DataFrame(data, columns=["Link", "PageRank"])

# Charger les données
df = load_data(r"C:\Users\Admin\Documents\GitHub\PageRank2024\part.txt")

# Fonction pour afficher le graphique interactif avec tri
def plot_interactive_bar_chart(score_range=(0.0, 0.6)):
    # Filtrer les données selon la plage sélectionnée
    min_value, max_value = score_range
    filtered_df = df[(df["PageRank"] >= min_value) & (df["PageRank"] <= max_value)]
    
    # Trier les données par ordre croissant de PageRank
    filtered_df = filtered_df.sort_values(by="PageRank", ascending=True)
    
    # Créer le graphique en barres
    fig = px.bar(filtered_df, x="Link", y="PageRank", title="Interactive Bar Chart of Links by PageRank Score (Sorted)")
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()

# Curseur pour la valeur maximale uniquement
interact(plot_interactive_bar_chart, score_range=FloatRangeSlider( min=0.0, max=0.6, step=0.01))
```
